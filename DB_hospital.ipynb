```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# 데이터프레임을 읽고 출력

hospital = pd.read_csv('동물병원정보.csv',encoding='utf-8')

hospital
```

```python
#NaN 없음으로 처리.
hospital['소재지전화'] = hospital['소재지전화'].fillna(value="없음")
hospital['도로명전체주소'] = hospital['도로명전체주소'].fillna(value="없음")

#앞 뒤 공백 제거
hospital['소재지전화'] = hospital['소재지전화'].str.strip()
hospital['도로명전체주소'] = hospital['도로명전체주소'].str.strip()
hospital['사업장명'] = hospital['사업장명'].str.strip()

#unnamed 없애기
hospital=hospital.drop(['Unnamed: 0', 'Unnamed: 32'], axis=1)
hospital
```

```python
# 데이터프레임의 데이터타입과 값이 없는 행의 개수를 출력
print(hospital.dtypes)
print(hospital.isnull().sum())
```

```python
#전화번호
call = hospital['소재지전화'].str.extract('(\d\d\d-\d\d\d-\d\d\d\d)')
call = call.dropna(subset=[0])
call
```

```python
#DB, 사용자 선택목록 보여줄때 사용할 자료(?)
Location = hospital['도로명전체주소'].str.split(" ")
first = Location.str[0]
second = Location.str[1]

first.to_csv("시도")
second.to_csv("시군구")
```
