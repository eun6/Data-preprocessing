```python
# 데이터프레임을 읽고 출력

petDisease = pd.read_csv('petDiseaseDiseasePage.csv',encoding='utf-8', lineterminator='\n')

petDisease
```

```python
petDisease['보호자에 대한 조언'] = petDisease['보호자에 대한 조언'].str.replace("\r","")
petDisease['보호자에 대한 조언'] = petDisease['보호자에 대한 조언'].str.replace("  "," ")
petDisease['보호자에 대한 조언'] = petDisease['보호자에 대한 조언'].str.replace("# ","")
petDisease['보호자에 대한 조언'] = petDisease['보호자에 대한 조언'].str.replace("#","")

petDisease['예방'] = petDisease['예방'].str.replace("\r","")
petDisease['예방'] = petDisease['예방'].str.replace("  "," ")
petDisease['예방'] = petDisease['예방'].str.replace("# ","")
petDisease['예방'] = petDisease['예방'].str.replace("#","")
petDisease
```

```python
# 개/고양이로 나누는 프로세스
# 개를 포함하는 행만 추출해서 petDiseaseDog 데이터프레임에 저장
petDiseaseDog = petDisease.loc[petDisease.index[petDisease['축종'].str.contains('개')].tolist(),:]
```
